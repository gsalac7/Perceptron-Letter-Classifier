3.	Implement a perceptron to recognize the letter A by producing output +1 for images of the letter A and -1 for images of all other letters.
Begin with weights initialized to random values and a learning rate of 𝛼 = 0.01. Train the perceptron by applying the update rule of Equation 1.4 in the textbook. Continue training until all items in TRAINING_SET are correctly classified.


```python
import numpy as np

learning_rate = 0.01
n = 50 
bias = -5

# Add two vectors together
def add_list(list1, list2):
    result = []
    for i in range(len(list1)):
        result.append(list1[i] + list2[i])
    return result

# product of two vectors
def product(multiple, inputs):
    product = []
    for i in inputs:
        product.append(i * multiple)
    return product

# Update function taken from 1.4 from textbook
# weights = weights + learning_rate(actual_value - predicted_value) * inputs
def update_function(actual_value, predicted_value, weights, inputs):
    new_weights = product(learning_rate*(actual_value - predicted_value), inputs)
    new_weights = add_list(weights, new_weights)
    return new_weights
    
# Activation Function
def sign(val):
    if val > 0:
        return 1 # It is the letter A
    else:
        return -1 # It is NOT the letter A

# Run the guess function
# Sum of all inputs * weights
def guess(inputs, weights):
    sum = 0
    for i in range(len(inputs)):
        sum += inputs[i] * weights[i]
    sum += bias
    return sign(sum)

# Error function
def error(actual_value, predicted_value):
    err = actual_value - predicted_value
    return err

# run the algorithm to classify a letter
def run(letter):
    weights = np.random.rand(35)
    EPOCH = 1
    for i in range(n):
        print("EPOCH " + str(EPOCH))
        print("*****")
        print("Starting FONT 1 TRAINING: ")
        print("*****")
        EPOCH += 1
        for i in range(52):
            predicted_value = guess(TRAINING_SET[i][0], weights)
            actual_value = 0
            if TRAINING_SET[i][1] == letter:
                actual_value = 1
            else:
                actual_value = -1
            # the guess is incorret, need to fix
            if actual_value != predicted_value:
                weights = update_function(actual_value, predicted_value, weights, TRAINING_SET[i][0])

            if i == 26:
                print("*****")
                print("Starting FONT 2 TRAINING: ")
                print("*****")
            print("Predicted Value: " + str(predicted_value) + "   " + str(TRAINING_SET[i][1]) + " Actual Value: " + str(actual_value) + "   " + str(TRAINING_SET[i][1]))

# Run the perceptron to classify the letter "A"          
run("A")
```
